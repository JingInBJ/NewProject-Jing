```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('birth rate-WB.csv')
df.head()
```

```python
df_china = df[['Year','China']]
df_china = df_china.rename(columns={'China':'Birth_rate'})
df_china.head()
```

```python
df_china['Country'] = 'China'
df_china.head()
```

```python
df_japan = df[['Year','Japan']]
df_japan = df_japan.rename(columns={'Japan':'Birth_rate'})
df_japan['Country'] = 'Japan'
df_japan.head()
```

```python
df_korea = df[['Year','Korea']]
df_korea = df_korea.rename(columns={'Korea':'Birth_rate'})
df_korea['Country'] = 'Korea'
df_korea.head()
```

```python
df_middle = df[['Year','Middle_income']]
df_middle = df_korea.rename(columns={'Middle_income':'Birth_rate'})
df_middle['Country'] = 'Middle_income'
df_middle.head()
```

```python
frames = [df_china, df_japan, df_korea, df_middle]
result = pd.concat(frames)
```

```python
result
```

```python
result.to_csv("birth_rate-final.csv")
```
