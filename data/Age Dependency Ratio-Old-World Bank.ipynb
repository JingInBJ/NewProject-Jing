# 在生成的csv里用了行列转置，接着下面的操作。（应该有更好的办法）

```python
df1 = pd.read_csv('AgeDeRatio-WB.csv')
df1.head()
```

```python
df1 = df1.rename(columns={'Country Name':'Year'})
df1 = df1.rename(columns={'Korea, Rep.':'South Korea'})
df1.head()
```

```python
df1['China']=df1['China'].round(1)
df1['Japan']=df1['Japan'].round(1)
df1['South Korea']=df1['South Korea'].round(1)
df1.head()
```

```python
df_china = df1[['Year','China']]
df_china = df_china.rename(columns={'China':'AgeDeRatio'})
df_china.head()
```

```python
df_china['Country'] = 'China'
df_china.head()
```

```python
df_Japan = df1[['Year','Japan']]
```

```python
df_Japan.head()
```

```python
df_Japan = df_Japan.rename(columns={'Japan':'AgeDeRatio'})
df_Japan.head()
```

```python
df_Japan['Country'] = "Japan"
df_Japan.head()
```

```python
df_sk = df1[['Year','South Korea']]
df_sk = df_sk.rename(columns={'South Korea':'AgeDeRatio'})
df_sk['Country'] = "South Korea"
df_sk.head()
```

```python
frames = [df_china, df_Japan, df_sk]
result = pd.concat(frames)
result
```

```python
result.to_csv('AgeDeRatio-fordatawrapper.csv')
```
